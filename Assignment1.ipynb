```python
'''[Q1-answer:-]'''
Iterative statements - if we want to execute a group of statements multiple times then we should
              go for iterative statement.
python supports two(2) type of iterative statements.
    1. for loop
    2. while loop
    
(1) for loop :-if we want to execute some action for every element present in some sequence (it may be string
                or collection) then we should go for for loop.

syntax---
            for i in sequence:
                body
            
        where sequence can be string or any collection.
        body will be execute for every element present in the sequence
        
example - to print character present in the given string.

        string = 'pwskills'
        for i in string:
            print(i)
            
        output-
                p
                w
                s
                k
                i
                l
                l
                s
                
(2)while loop:- if we want to execute a group of statements iteratively until some condition false, then we
                should go for while loop.

syntax--
        while condition:
            body
            
example- To print the number till given input.
            
        number = int(input("enter the nummber :"))
        i = 1
        while i<=number:
            print(i," ",end='')
            i=i+1
        
        output--
                1 2 3 4 5 6 7 8 9 .... number
    
```

```python
'''[Q2-answer:-]'''
#To print sum and product of the first 10 natural number using for loop
number = 10
sum = 0
product=1
for i in range(1,number+1):
    sum+=i
    product*=i
print("sum of number = ",sum)
print("product of number = ",product)

#To print sum and product of the first 10 natural number using while loop
number1 = 10
i = 1
sum1 = 0
product1 = 1
while i<=number1:
    sum1+=i
    product1*=i
    i+=1
print("sum of number = ",sum1)
print("product of number = ",product1)
```

```python
'''[Q4-answer:-]'''
#using for loop to calculate number cubes which is divisible by 4 and 5 
n=int(input("enter the 100 number :"))
x,y = [],[]
for i in range(1,n+1):
    y.append(i)
print("input number from 1 to 100 in list = ",y)

for i in y:
    z=i**3
    if z%4==0 or z%5==0:
        x.append(z)
print("divisible number cube by 4 and 5 = ",x)

#using while loop to calculate number cubes which is divisible by 4 and 5 
n1=int(input("enter the 100 number :"))
a,b= 0,1
r,p = [],[]
while b <= n1:
    p.append(b)
    b+=1
print("input number from 1 to 100 in list = ",p)
while a<n1:
    w = p[a]**3
    if w%4==0 or w%5==0:
        r.append(w)
    a+=1
print("divisible number cube by 4 and 5 = ",r)
```
